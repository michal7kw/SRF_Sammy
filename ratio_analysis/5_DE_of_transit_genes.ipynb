```python
import pandas as pd

dea_neu_path = "/beegfs/scratch/ric.broccoli/kubacki.michal/SRF_MeCP2_CUTandTAG/DATA/DEA_NEU.csv"
dea_neu = pd.read_csv(dea_neu_path)

dea_neu.head()
```

```python
dea_nsc_path = "/beegfs/scratch/ric.broccoli/kubacki.michal/SRF_MeCP2_CUTandTAG/DATA/DEA_NSC.csv"
dea_nsc = pd.read_csv(dea_nsc_path)

dea_nsc.head()
```

## NSC_GFP_vs_NSC_M2_B_to_A_genes

```python
nsc_b_to_a_genes_path = "/beegfs/scratch/ric.broccoli/kubacki.michal/SRF_Sammy/ratio_analysis/results/shift_GFP_M2_with_blacklist_with_mirko_gene_lists/shifted_regions/gene_associations/NSC_GFP_vs_NSC_M2_B_to_A_genes.tsv"
nsc_b_to_a_genes_df = pd.read_csv(nsc_b_to_a_genes_path, sep='\t')

nsc_b_to_a_gene_names = nsc_b_to_a_genes_df['gene_name'].tolist()
print(nsc_b_to_a_gene_names)
```

```python
dea_nsc_transit = dea_nsc[dea_nsc['gene'].isin(nsc_b_to_a_gene_names)]

dea_nsc_transit.sort_values(by='padj').head()
```

# <img src="2025-03-12 08_43_34-IGV.png">


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style - using a valid matplotlib style
plt.style.use('default')  # Or try 'ggplot', 'seaborn-v0_8', etc.

plt.figure(figsize=(12, 8))

# Create -log10(padj) column
dea_nsc_transit.loc[:, '-log10padj'] = -np.log10(dea_nsc_transit['padj'])

# Define significance thresholds
alpha = 0.05
log2fc_threshold = 1

# Create significance categories
dea_nsc_transit.loc[:, 'regulation'] = 'Not Significant'
dea_nsc_transit.loc[(dea_nsc_transit['padj'] < alpha) & (dea_nsc_transit['log2FoldChange'] > log2fc_threshold), 'regulation'] = 'Upregulated'
dea_nsc_transit.loc[(dea_nsc_transit['padj'] < alpha) & (dea_nsc_transit['log2FoldChange'] < -log2fc_threshold), 'regulation'] = 'Downregulated'

# Color scheme
colors = {'Upregulated': '#FF4B4B', 'Downregulated': '#4B4BFF', 'Not Significant': '#CCCCCC'}

# Create scatter plot
scatter = sns.scatterplot(
    data=dea_nsc_transit,
    x='log2FoldChange',
    y='-log10padj',
    hue='regulation',
    palette=colors,
    s=50,
    alpha=0.7,
    edgecolor='none'
)

# Add threshold lines
plt.axhline(y=-np.log10(alpha), color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=log2fc_threshold, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=-log2fc_threshold, color='gray', linestyle='--', alpha=0.5)

# Customize plot
plt.title('Differential Expression Analysis of NSC Transit Genes', pad=20, size=14, weight='bold')
plt.xlabel('log₂ Fold Change', size=12)
plt.ylabel('-log₁₀ Adjusted p-value', size=12)

# Adjust legend
plt.legend(title='Gene Regulation', bbox_to_anchor=(1.05, 1), loc='upper left')

# Add grid
plt.grid(True, linestyle='--', alpha=0.3)

# Adjust layout
plt.tight_layout()

plt.show()
```
